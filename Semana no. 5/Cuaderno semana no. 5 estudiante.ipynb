### Matplotlib es una biblioteca de Python fundamental para la visualización de datos, permitiendo crear gráficos estáticos, animados e interactivos en 2D (y 3D) con alta calidad. Diseñada para integrarse con NumPy y Pandas, facilita la creación de diagramas de dispersión, barras, histogramas, entre otros.

### Entre sus principales características tenemos:

### Funcionalidad: Se puede personalizar (colores, etiquetas, etc.)
### Versatil: Puede aceptar fotografias
### Uso variado: sobre todo en análisis exploratorio de datos.

### Muchas veces puede unirse con la librería Seaborn, para lograr una mejor comprensión de la data.
