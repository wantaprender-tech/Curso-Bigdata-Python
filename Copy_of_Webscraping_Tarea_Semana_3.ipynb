```python
#cargar libreria
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Asignamos el sitio Web, y luego obtenemos la info
url="https://www.scrapethissite.com/pages/simple/"
info=requests.get(url)
#Validamos si fue exitosa la funcion
info
```

```python
#Usaremos beautifulsoup para dar forma y entender las etiquetas con el atributo html.parser
soup=BeautifulSoup(info.text,'html.parser')
print(soup)
```

```python
#Guardamos los paises bajo el nombre countries
countries = soup.select('.country')
```

```python
#Haremos un ciclo para extraer todos los nombres, poblacion y area. Observamos en el paso anterior que los nombres de la poblacion y area estan en:
#country-population y country-area

data = []

for c in countries:
    nombre = c.select_one('.country-name').get_text()
    poblacion = c.select_one('.country-population').get_text()
    area = c.select_one('.country-area').get_text()

    data.append([nombre, poblacion, area])

data
```

```python
#Ahora meteremos esto en un DataFrame
df = pd.DataFrame(data, columns=['pais', 'poblacion', 'area'])
df
```

```python
#Vamos a limpiar la data en la columna de población y convertirla a números y la de pais, que tiene valores no adecuados como "\n"
#Iniciemos con quitar espacios y saltos de linea
df['pais'] = df['pais'].str.strip()
```

```python
#Ahora quitamos espacios y slatos de linea en plblacion y eliminamos las comas de miles
df['poblacion'] = df['poblacion'].str.strip()
df['poblacion'] = df['poblacion'].str.replace(',', '', regex=False)

df['poblacion'] = pd.to_numeric(df['poblacion'], errors='coerce').astype('Int64')
''
```

```python
#validamos que poblacion este como numero int
df.head()
df.dtypes
```

```python
#Revisemos el Data Frame
df
```

```python
#Determinando la población total del mundo según este archivo
poblacion_mundial = df['poblacion'].sum()
print(f"La población mundial es de {poblacion_mundial} habitantes.")
```

```python
#Como area es tipo object, debemos convertirlo a float

df['area'] = (
    df['area']
      .str.strip()
      .str.replace(',', '', regex=False)
      .pipe(pd.to_numeric, errors='coerce')
)
```

```python
#validamos que area este como float
df.head()
df.dtypes
```

```python

#Promedio de área en km2 por país con 2 decimales
promedio_area = df['area'].mean()
print(f"El promedio de área por país es de {promedio_area:.2f} km2.")
```

```python
#Población total de Centroamérica (incluyendo Belice)
paises_centroamerica = [
    'Belize', 'Guatemala', 'Honduras', 'El Salvador',
    'Nicaragua', 'Costa Rica', 'Panama'
]
poblacion_centroamerica = (df[df['pais'].isin(paises_centroamerica)]['poblacion'].sum())
print(f"La población total de Centroamérica es de {poblacion_centroamerica} habitantes.")
```

```python
#Porcentaje de la población centroamericana respecto al resto del mundo con 2 decimales
resto_mundo = (poblacion_mundial - poblacion_centroamerica) if pd.notna(poblacion_mundial) else None
porcentaje_ca_vs_resto = (poblacion_centroamerica / resto_mundo * 100) if (resto_mundo and resto_mundo > 0) else None
print(f"El porcentaje de la población Centroamérica respecto al resto del mundo es de {porcentaje_ca_vs_resto:.2f}%.")
```
