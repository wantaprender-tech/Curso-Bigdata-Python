## Archivos Json:

Un archivo JSON (JavaScript Object Notation) es un formato de texto ligero y estándar para el almacenamiento e intercambio de datos, diseñado para ser fácil de leer por humanos y analizable por máquinas. Se basa en pares de clave-valor y se utiliza ampliamente en aplicaciones web para transferir información entre servidores y aplicaciones.

Estructura: Se organiza mediante objetos (colecciones de pares {"clave": "valor"} encerradas en llaves {}) y arreglos (listas ordenadas de valores encerradas en corchetes [])

¿Para qué sirve?
JSON es el estándar principal para APIs web, configuración de aplicaciones, almacenamiento de datos no estructurados en bases de datos NoSQL (como MongoDB) y la transferencia de información en tiempo real.

Ahora, vamos a cargar un documento desde una API, esto se realiza con la función requests

Una API (Interfaz de Programación de Aplicaciones) es un conjunto de reglas, protocolos y herramientas que permite que dos aplicaciones de software se comuniquen e intercambien datos entre sí. Actúa como un intermediario, facilitando que una aplicación utilice funciones o recursos de otra, como mostrar el clima, pagos o mapas, sin necesidad de desarrollarlos desde cero.

Una API es como un menú, donde usted pide algo a una dirección (URL) y el servidor devuelve la información generalmente en un archivo JSON.

la dirección de nuestra API es: https://jsonplaceholder.typicode.com/users

```python
import requests
url2="https://jsonplaceholder.typicode.com/users"
respuesta=requests.get(url2)
df_json=respuesta.json()
df_json
```

Con esta instrucción, tenemos el archivo Json en pantalla, es decir, los datos, pero debemos de transformarlo en una tabla para poder analizar que data existe, para ello usamos el comando DataFrame de pandas

```python
df_jsonf=pd.DataFrame(df_json)
df_jsonf
```

Como pueden ver, la columna Address y la columna Company tienen más datos adentro (anidadas), a esto dentro de los archivos Json se le conoce como diccionarios (2 más columnas anidadas dentro de una misma). 

Para este proceso hay que aplanar el archivo de forma que python separe los datos anidados, puede hacerse de dos formas, utilizando la instrucción json_normalize (que volverá listas las columnas anidadas, esta implica volver a cargar el archivo) y la función tolist(), que cuando ya tenemos cargado el archivo nos ayuda a hacer este proceso.



```python
json_address=pd.json_normalize(df_jsonf['address'].tolist())
json_address=json_address.add_prefix('addr')
json_address.head()
```

```python
df_jsonf.head()
```

Ahora, haremos lo mismo pero con la compañía

```python
json_company=pd.json_normalize(df_jsonf['company'].tolist())
json_company=json_company.add_prefix('comp_')
json_company.head()
```

A veces, sucede un error cuando la columna tiene más de un diccionario (la forma de un diccionario es [{}], a veces puede venir por ejemplo alguna columna nula o [[{}]], y eso da error), para ello vamos a utilizar otra forma de hacerlo:


```python
#lista_company=list(df_jsonf['company'])
#json_company=pd.json_normalize(lista_company)
#json_company=json_company.add_prefix('comp_')
#json_company.head()
```

Ahora, hay que unir los datos que se arreglaron al dataframe original, para ello usaremos la función concat, y eliminaremos las columnas que ya no sirven.

**Ojo, recordar copiar o crear un nuevo dataframe antes de hacer cualquier cambio por si cometemos un error.
Ojo, es importante notar que la función concat requiere una instrucción llamada Axis, siendo 0 para llenar por filas, siendo 1 para llenar por columnas.**

```python
df_jsonfinal=pd.concat([df_jsonf,json_address,json_company], axis=1)
df_jsonfinal=df_jsonfinal.drop(['address','company'],axis=1)
df_jsonfinal.head()

```

# ahora, vamos a cambiar a SQL
## con sql, vemos que son datos estructurados, tienen relaciones y están establecidos en una base de datos de acceso restringido (link, usuario, clave, puerto), lo cual requiere un proceso de seguridad en la conexión.

### nosotros, vamos a usar sqlite para simular una conexión de sql, con el fin de evitar tener que montar un servidor sql.  

### igualmente, nos conectaremos a una base de datos sql en web pero que es sitio seguro, con el fin de buscar también el trámite de información, pero no requeriremos credenciales.


### la libreria a utilizar en python para sql es sqlite 3


```python
import sqlite3
```

```python
# para este primer ejemplo, vamos a crear una database local pequeña, que usaremos como conexión tipo sql, con el fin
# de hacer nuestra primera conexión.

datos ={
    'ID proyecto':[101,102,103,104,105,106,107],
    'Nombre': ['Proyecto 1','Proyecto 2','Proyecto 3','Proyecto 4','Proyecto 5','Proyecto 6','Proyecto 7'],
    'Presupuesto': [50000,25000,80000,450000,75000,20000,125000],
    'Estado':["Completado","Completado","En curso","Pendiente","Pendiente","En curso","Pendiente"]
}
# asignamos la data a un dataframe con la función DataFrame
# DataFrame crea una estructura de datos bidimensional, etiquetada y tabular
# similar a una hoja de cálculo o tabla SQL, organizada en filas y columnas

df_proyectos=pd.DataFrame(datos)
df_proyectos

```

```python
# ahora, vamos a crear la conexión para simular un archivo sql para manipular la data con las instrucciones de sql
# hay dos instrucciones, if_exists (esta tiene tres status: append (agregar), fail(fallar) y replace(reemplazar)), y la
# función index sirve para decirle a python si agrega el index de el o solamente la data nuestra (True=agregar, False= no)

conexion=sqlite3.connect("GestionProyectos.db")
df_proyectos.to_sql('proyectos',conexion, if_exists='replace',index=False)
print("hecho")
```

```python
# ya que tenemos este proceso, ya podemos trabajar los comandos de sql (Select, from, where, etc.), TODOS LOS COMANDOS
# DE SQL SON EN MAYÚSCULAS
# usaremos una variable para almacenar la consulta que queremos

consulta="SELECT * FROM proyectos WHERE Presupuesto > 40000 AND Estado ='En curso'"


# ahora, leemos la consulta.
df_reporte=pd.read_sql(consulta,conexion)
df_reporte
```

```python
consulta="SELECT * FROM proyectos WHERE Presupuesto >40000"
df_reporte=pd.read_sql(consulta,conexion)
df_reporte
```
