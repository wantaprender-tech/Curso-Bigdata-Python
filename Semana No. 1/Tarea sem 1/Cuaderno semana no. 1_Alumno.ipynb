## Datos Estructurados vs. No Estructurados

En el mundo del Big Data, la primera gran clasificaci칩n de la informaci칩n depende de su organizaci칩n:

* **Datos Estructurados:** Son datos que tienen un formato fijo y bien definido (como una tabla de Excel o una base de datos SQL). Cada columna tiene un nombre y un tipo de dato (n칰mero, fecha, texto). 

    * *Ejemplo:* Una lista de clientes con Nombre, Tel칠fono y Ciudad.

* **Datos No Estructurados:** Informaci칩n que no tiene un formato predefinido. Es m치s dif칤cil de procesar autom치ticamente porque no hay filas ni columnas claras.

    * *Ejemplo:* Correos electr칩nicos, archivos de audio, PDFs, im치genes o mensajes de redes sociales.

* **Datos Semi-estructurados:** No son tablas perfectas, pero tienen etiquetas para organizar la informaci칩n.

    * *Ejemplo:* Archivos **XML** y **JSON**.

##### Es importante notar que la librer칤a pandas (la cual usaremos para la lectura de datos) es muy buena especialmente trabajando con datos **estructurados** y **semi-estructurados**.

## 3. Pandas: El motor de lectura de Python

Pandas no es solo para hacer c치lculos; es una herramienta de **Carga de Datos**. 

Cuando usamos funciones como `pd.read_csv()`, Pandas realiza un proceso llamado "Parsing" (o an치lisis sint치ctico), el cual es el proceso inform치tico de convertir datos desordenados, texto plano o c칩digo fuente en un formato estructurado y organizado. Es esencial para interpretar informaci칩n, extraer datos espec칤ficos, o analizar el significado gramatical y l칩gico de una entrada.

## 4. Formatos Principales de carga:

Para evitar errores de lectura, debemos entender c칩mo est치n construidos los archivos por dentro:

### A. El CSV (Comma Separated Values)

Es texto plano. Cada l칤nea es una fila y cada columna se separa por un signo (usualmente una coma `,` o punto y coma `;`).

**Importante: Si tu archivo usa `;` y t칰 le dices a Pandas que busque `,`, los datos se ver치n amontonados en una sola columna.


### B. El Excel (.xlsx)

Es un formato binario propietario de Microsoft. A diferencia del CSV, un archivo de Excel puede contener m칰ltiples **hojas** (Sheets).

**Importante: En Pandas, debemos especificar qu칠 hoja queremos leer usando el par치metro `sheet_name`.


### C. El XML (eXtensible Markup Language)

Es un formato basado en **jerarqu칤as y etiquetas** (parecido al HTML). No es una tabla plana, sino un "치rbol" de datos.

**Anatom칤a:** Se compone de **Nodos Ra칤z** y **Nodos Hijos**.

## Los componentes clave de la estructura XML incluyen: 

**Declaraci칩n XML (Pr칩logo): Es la primera l칤nea, define la versi칩n y codificaci칩n.**

**Elemento Ra칤z (Root): Es el contenedor principal que engloba a todos los dem치s elementos. Todo archivo XML debe tener exactamente un 칰nico elemento ra칤z.**

**Elementos (Tags): Definen la estructura jer치rquica. Tienen una etiqueta de apertura**

**Atributos: Proporcionan informaci칩n adicional sobre un elemento dentro de la etiqueta de apertura.**
    
**Contenido: Los datos reales situados entre las etiquetas de apertura y cierre.**
    
**Comentarios: Notas de c칩digo no procesadas**
