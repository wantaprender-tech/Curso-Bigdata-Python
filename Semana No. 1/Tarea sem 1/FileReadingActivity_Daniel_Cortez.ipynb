```python
import pandas as pd
```

```python
#cargando archivo csv
url_csv="https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv"
df_csv=pd.read_csv(url_csv)
```

```python
#Mostrando los primeros 6 registros
df_csv.head(6)
```

```python
#Mostrando la informacion del archivo
df_csv.info()
```

```python
#Tamaño del archivo
df_csv.shape
```

```python
#Cantidad de paises diferentes
# Agrupamos por pais
paises=df_csv.groupby("pais")["pais"].count()
paises.count()
```

```python
#Media de corrupción por año
df_csv.groupby("anio")["percepcion_corrupcion"].mean()
```

```python
#Expectativa de vida por pais
df_csv.groupby("pais")["expectativa_vida"].mean()
```

```python
#Archivo XML, lo cararemos como url_xml
url_xml="https://www.w3schools.com/xml/cd_catalog.xml"
df_xml=pd.read_xml(url_xml)
```

```python
#Mostramos los primeros 6 registros
df_xml.head(6)
```

```python
#Mostramos la informacion del archivo
df_xml.info()
```

```python
#Tamaño del archivo
df_xml.shape
```

```python
#Filtramos los discos entre 1990 y 1997
df_xml_filtrado=df_xml[(df_xml["YEAR"] >= 1990) & (df_xml["YEAR"] <= 1997)]
df_xml_filtrado
```

```python
#Disco mas viejo al mas reciente usando iloc con un db llamado df_xml_ordenado
df_xml_ordenado=df_xml.sort_values("YEAR")
#Mas viejo
df_xml_ordenado.iloc[0]
```

```python
#Mas reciente
df_xml_ordenado.iloc[-1]
```

```python
#Cargamos el archivo excel como df_excel
url_excel="https://github.com/wantaprender-tech/Curso-Bigdata-Python/raw/refs/heads/main/Semana%20No.%201/base%20de%20datos%20excel%20tarea%20sem1.xlsx"
df_excel=pd.read_excel(url_excel)
```

```python
#Mostramos los primeros 6 registros
df_excel.head(6)
```

```python
#Mostramos la información del archivo
df_excel.info()
```

```python
#Mostramos el tamaño del archivo
df_excel.shape
```

```python
# Realizamos una copia del df original y eliminaremos la columna cliente
df_excel_copia=df_excel.copy()
df_excel_copia=df_excel_copia.drop(columns=['Producto','Cliente'])
#Primeros 6 registros
df_excel_copia.head(6)
```

```python
#Ventas por vendedor
ventas_vendedor=df_excel_copia.groupby('Empleado')['Ventas'].sum()
ventas_vendedor
```

```python
# Creamos totales por categoría
total_categoria=df_excel_copia.groupby('Categoría')['Ventas'].sum()

# Asignamos valores máximos y mínimos
categoria_min=total_categoria.idxmin()
categoria_max=total_categoria.idxmax()

# Resultados
print(categoria_min)
print(categoria_max)
```
