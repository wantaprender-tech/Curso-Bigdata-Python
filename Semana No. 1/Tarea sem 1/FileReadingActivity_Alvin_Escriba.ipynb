```python
#Llamando a la libreria Panda
import pandas as pd
```

```python
#Trabajando el archivo CSV, y se va a llamar Tarea1_CSV. Luego lo cargamos a Dataframe 1 llamado df1
Tarea1_CSV="https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv"
df1=pd.read_csv(Tarea1_CSV)
```

```python
#Mostrando la información del archivo
df1.info()
```

```python
#Mostrando el tamaño del archivo (filas y columnas)
df1.shape
```

```python
#En los archivos independientes mostraremos cuantos países diferentes hay en la db
df1['pais'].value_counts()
#Encontramos valores vacios
```

```python
#Vamos a mostrar por año la media de corrupción
df1.groupby('anio')['percepcion_corrupcion'].mean()
```

```python
#Vamos a mostrar por país, la expectativa de vida
df1.groupby('pais')['expectativa_vida'].mean()
```

```python
#Trabajando el archivo XML, y se va a llamar Tarea1_XML. Luego lo cargamos a Dataframe 2 llamado df2
Tarea1_XML="https://www.w3schools.com/xml/cd_catalog.xml"
df2=pd.read_xml(Tarea1_XML)
```

```python
#Mostrando el encabezado de los primeros 6 registros
df2.head(6)
```

```python
#Mostrando la información del archivo
df2.info()
```

```python
#Mostrando el tamaño del archivo (filas y columnas)
df2.shape
```

```python
#Filtraremos los discos para que muestre aquellos que esten entre el año 1990 y 1997
df2[(df2['YEAR']>=1990) & (df2['YEAR']<=1997)]
```

```python
#vamos a mostrar con la función iloc cual es el disco más viejo y el más reciente. Primero creamos una copia de trabajo llamado YEAR_temp, luego sorteamos los archivos por el campo YEAR por orden ascendente, y utilizamos la funcion inplace
YEAR_temp=df2.copy()
YEAR_temp.sort_values(by='YEAR',ascending=True,inplace=True)
```

```python
#Ahora ya podemos separar el valor mas nuevo (mas_nuevo) y mas antiguo (mas_antiguo) utilizando las funciones iloc y despues los desplegamos
mas_antiguo=YEAR_temp.iloc[[0]]
mas_nuevo=YEAR_temp.iloc[[-1]]
print(mas_antiguo[['TITLE','ARTIST','COUNTRY','COMPANY','PRICE','YEAR']])
print(mas_nuevo[['TITLE','ARTIST','COUNTRY','COMPANY','PRICE','YEAR']])

```

```python
#Trabajando el archivo XLSC, y se va a llamar Tarea1_XLSX. Luego lo cargamos a Dataframe 1 llamado df3
Tarea1_XLSX="https://github.com/wantaprender-tech/Curso-Bigdata-Python/raw/refs/heads/main/Semana%20No.%201/base%20de%20datos%20excel%20tarea%20sem1.xlsx"
df3=pd.read_excel(Tarea1_XLSX)
```

```python
#Mostrando el encabezado de los primeros 6 registros
df3.head(6)
```

```python
#Mostrando la información del archivo
df3.info()
```

```python
#Mostrando el tamaño del archivo (filas y columnas)
df3.shape
```

```python
#Indique si hay columnas que borrar (cuales son?) Al observar en los headers, encontramos valores duplicados Producto, Nombre del producto, Cliente y Nombre del cliente.Se pueden conservar los nombres mas explicitos
#Nombre del cliente y Nombre del producto.
```

```python
#Realizaremos una copia de su dataframe y en este deje solo los datos que considere necesarios, es decir que borramos Produto y Cliente. Luego mostraremos nuevamente los primeros 6 registros
df3_limpio=df3.copy()
df3_limpio.drop(columns=['Producto','Cliente'])
df3_limpio.head(6)
```

```python
#Muestre el total de ventas por vendedor.
df3_limpio.groupby('Empleado')['Ventas'].sum()
```

```python
# Vamos a mostrar la categoría que más vende y la que menos vende. Para ello, creamos totales por categoria en total_categoria
total_categoria=df3_limpio.groupby('Categoría')['Ventas'].sum()

```

```python
#Ahora asignamos valores maximos y minimos en los campos categoria_min y categoria_max
categoria_min=total_categoria.idxmin()
categoria_max=total_categoria.idxmax()
```

```python
#Luego mostramos los valores de categoria_max y categoria_min
print(categoria_min)
print(categoria_max)
```
