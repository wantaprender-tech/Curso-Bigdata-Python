# Primero lo primero, que es una API

## Una api (Interfaz de programación de aplicaciones) es un conjunto de reglas y protocolos que permiten que diferentes aplicaciones de software se comuniquen e intercambien datos entre sí de forma automática.

## Básicamente es un puente de conexión entre dos sistemas

### Estas utilizan solicitudes y respuestas a través de los protocolos Http y Https, permitiendo que un cliente solicite información a un servidor.

### Es muy importante tomar en cuenta lo siguiente:

### Un api tiene documentación, es importante leer su documentación para que podamos entender como trabaja, que protocolos de comunicación usa y cuales son los comandos que puede utilizar para traer información.

### Toda api tiene puntos finales (endpoints), que básicamente son las URL donde envia los datos y solicitudes

### En la documentación normalmente están los métodos de acceso (get, put, etc.), estos no todos son estandar

### La mayoría de respuestas están en formato Json o bien XML

### Hay api pública y privada, las públicas normalmente no le piden usuario o llave (key), mientras que las privadas si lo requieren
