### En el lenguaje HTML lo primero que hay que comprender es que todo está relacionado a etiquetas, las cuales tienen jerarquía.  Estas etiquetas suelen conocerse con el nombre de Tags o contenedores, se reconocen por estar escritas <tag> y </tag>

### También existen los atributos, los cuales son información que existe dentro de una etiqueta.  Para nuestro uso, son las clases (class_, identifica grupos) e identificadores (id, identifica elementos únicos).

|  Etiqueta  |  Significado  |  Uso en Scraping  |
|  :---  |  :---  |  :---  |
| **h1 a h6**  |  Títulos  |  Nombres de productos o artículos  |
|  **p**  |  Párrafo  |  Descripciones o precios  |
|  **a**  |  Enlace (Anchor)  |  Para obtener la URL con el atributo href  |
|  **img**  |  Imagen  |  Para obtener la foto con el atributo src  |
|  **ul> / <li**  |  Listas  |  Menús o características técnicas  |
